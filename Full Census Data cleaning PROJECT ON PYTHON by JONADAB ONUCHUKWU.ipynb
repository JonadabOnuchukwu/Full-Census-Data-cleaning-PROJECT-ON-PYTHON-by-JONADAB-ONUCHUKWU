```python
def employment_status(role):
    if role == 'Child':
        return 'Minor'
    elif role =='Unkown':
        return role
    elif 'Student' in role:
        return 'Student'
    elif 'Retired'in role:
        return 'Retired'
    else:
        return 'Employed'
```

```python
census_df_copy['employment_status'] = census_df_copy['Occupation'].apply(employment_status)
```
